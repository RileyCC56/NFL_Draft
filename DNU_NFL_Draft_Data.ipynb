```python
# Import scraping modules
from urllib.request import urlopen
from bs4 import BeautifulSoup
```

```python
# Import data manipulation modules
import pandas as pd
import numpy as np

# Import data visualization modules
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
# URL of page
url = 'https://stathead.com/football/draft_finder.cgi?request=1&show=all&draft_slot_min=1&draft_slot_max=500&draft_year_max=2021&order_by=default&draft_positions[]=qb&draft_positions[]=rb&draft_positions[]=wr&draft_positions[]=te&draft_positions[]=e&draft_positions[]=t&draft_positions[]=g&draft_positions[]=c&draft_positions[]=ol&draft_positions[]=dt&draft_positions[]=de&draft_positions[]=dl&draft_positions[]=ilb&draft_positions[]=olb&draft_positions[]=lb&draft_positions[]=cb&draft_positions[]=s&draft_positions[]=db&draft_year_min=2000&order_by_asc=0&conference=any&pick_type=overall&offset=1500#results'

# Open URL and pass to BeautifulSoup
html = urlopen(url)
draft_page = BeautifulSoup(html)
```

```python
# Collect table headers
column_headers = draft_page.findAll('tr')[1]
column_headers = [i.getText() for i in column_headers.findAll('th')]
```

```python
print(column_headers)
```

```python
# Collect table rows
rows = draft_page.findAll('tr')[1:]

# Get stats from each row
qb_stats = []
for i in range(len(rows)):
    qb_stats.append([col.getText() for col in rows[i].findAll('td')])
```

```python
# Create DataFrame from our scraped data
data = pd.DataFrame(qb_stats, columns=column_headers[1:])
```
